```python
# Generate product data
def generate_product_data(num_rows):
    product_data = []
    for _ in range(num_rows):
        product_name = fake.word() + ' Pipe'
        diameter = fake.random_int(min=1, max=24)  # Random diameter between 1 and 24 inches
        material = fake.word()  # Fake material name
        price = round(fake.random_number(digits=2), 2)  # Random price with 2 decimal places
        supplier_id = random.randint(1001, 1050)  # Random supplier ID
        category = fake.word()  # Fake category
        product_data.append({'Product Name': product_name, 'Diameter (inches)': diameter,
                             'Material': material, 'Price ($)': price, 'Supplier_ID': supplier_id,
                             'Category': category})
    return product_data

# Generate product data for 500 rows
num_rows = 500
product_data = generate_product_data(num_rows)

# Create DataFrame
product_df = pd.DataFrame(product_data)

# Display DataFrame
print(product_df.head())
```
